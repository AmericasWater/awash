In runs with constant (non-optimized) water demand, we apply water demands from USGS 2010 county-level demand estimates.

We also divide irrigation water across months, according to planting and harvesting dates from https://nelson.wisc.edu/sage/data-and-models/crop-calendar-dataset/crop-readme.txt, and crop planted area from 2010. 

This is constructed first by developing unnormalized weights, which assume that irrigation needs are constant per hectare over the entire period that a crop is planted:
$$\hat{w}_d = \sum_c a_c p_{cd}$$
where $a_c$ is the area of planted with crop $c$, and $p_{cd}$ is an indicator for whether the day $d$ is between the planting and harvesting dates for crop $c$. If no crop calendar is available, $p_{cd} = 1$ for all $d$. If multiple crop calendars are available, typically for a main season and a second season, then $p_{cd} = \frac{2}{3} p_{cd}^{\text{main}} + \frac{1}{3} p_{cd}^{\text{2nd}}$.

Then we construct a weight that sums to 1:
$$w_d = \frac{\hat{w}_d}{\sum_d \hat{w}_d}$$
